```python
import os
os.chdir('..')
import glob
import numpy as np
import tensorflow as tf
import config_root
from utils.geotif_io import readTiff
from dataloader.path_io import crop_patch
from dataloader.tfrecord_io import image_example
```

```python
path_tfrecord_scene = 'data/tfrecord-l578/l578_scene.tfrecords'

truth_paths = sorted(glob.glob(config_root.dset_l578_truth+'/*.tif'))
scene_paths = [config_root.dset_l578_scene + '/' + os.path.split(path)[-1].replace('_truth','') for path in truth_paths]
pair_data = list(zip(scene_paths, truth_paths))
print('data length:', len(pair_data))
print(pair_data[0])
```

```python
# Trainging data: Write to a `.tfrecords` file.
with tf.io.TFRecordWriter(path_tfrecord_scene) as writer:
    for path_scene, path_truth in pair_data:
        print('scene written:' ,path_scene)
        scene,_ = readTiff(path_scene)
        truth,_ = readTiff(path_truth)
        scene = np.clip(scene/10000,0,1)  
        tf_example = image_example(scene, truth)
        writer.write(tf_example.SerializeToString())
```
